```python
# Add Matplotlib inline magic comand
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# files to load

city_data_to_load = 'Resources/city_data.csv'
ride_data_to_load = 'Resources/ride_data.csv'
```

```python
# read city data into data frame

city_data_df = pd.read_csv(city_data_to_load)
city_data_df
```

```python
# read ride date into df

ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df
```

```python
# cehcking for complete city data
city_data_df.count()
```

```python
# Get the unique values of teh type of city
city_data_df['type'].unique()
```

```python
# Get the number of Urban city data points
sum(city_data_df['type'] == 'Urban')
```

```python
# Get the number of Suburban city data points
sum(city_data_df['type'] == 'Suburban')
```

```python
# Get the number of Rural city data points
sum(city_data_df['type'] == 'Rural')
```

```python
# inspecting ride data df - get column and rows that are not null
ride_data_df.count()
```

```python
# ensure no null values
ride_data_df.isnull().sum()
```

```python
# check data types
ride_data_df.dtypes
```

```python
# merging data frames
pyber_data_df = pd.merge(ride_data_df, city_data_df, how='left', on=['city', 'city'])

pyber_data_df.head()
```

```python
# Create Urban city dataframe

urban_cities_df = pyber_data_df[pyber_data_df['type'] == 'Urban']
urban_cities_df.head()
```

```python
# Create Suburban City DF

suburban_cities_df = pyber_data_df[pyber_data_df['type'] == 'Suburban']
suburban_cities_df.head()
```

```python
# Create Rural city DF
rural_cities_df = pyber_data_df[pyber_data_df['type'] == 'Rural']
rural_cities_df.head()
```

```python
# get the number of rides for urban cities

urban_ride_count = urban_cities_df.groupby(['city']).count()['ride_id']
urban_ride_count.head()
```

```python
# get the number of rides for suburban cities

suburban_ride_count = suburban_cities_df.groupby(['city']).count()['ride_id']
suburban_ride_count.head()
```

```python
# get the number of rides for rural cities
rural_ride_count = rural_cities_df.groupby(['city']).count()['ride_id']
rural_ride_count.head()
```

```python
# get the average far for urban cities
urban_avg_fare = urban_cities_df.groupby(['city']).mean()['fare']
urban_avg_fare.head()
```

```python
# get the average fare for suburban cities
suburban_avg_fare = suburban_cities_df.groupby(['city']).mean()['fare']
suburban_avg_fare.head()
```

```python
# get the average far for rural cities
rural_avg_fare = rural_cities_df.groupby(['city']).mean()['fare']
rural_avg_fare.head()
```

```python
# get the average number of drivers in urban cities
#### only following module directions - this step seems unnecessary since the number of drivers for a given city doesn't change - no need to take the average

urban_driver_count = urban_cities_df.groupby(['city']).mean()['driver_count']
urban_driver_count.head()
```

```python
# get the average number of drivers in suburban cities
#### only following module directions - this step seems unnecessary since the number of drivers for a given city doesn't change - no need to take the average

suburban_driver_count = suburban_cities_df.groupby(['city']).mean()['driver_count']
suburban_driver_count.head()
```

```python
# get the average number of drivers in rural cities
#### only following module directions - this step seems unnecessary since the number of drivers for a given city doesn't change - no need to take the average

rural_driver_count = rural_cities_df.groupby(['city']).mean()['driver_count']
rural_driver_count.head()
```

```python
# Build the scatter plots for urban cities.
plt.scatter(urban_ride_count,
      urban_avg_fare,
      s=[i*10 for i in urban_driver_count], c="coral",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Urban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
# Build the scatter plots for suburban cities.
plt.scatter(suburban_ride_count,
      suburban_avg_fare,
      s=[i*10 for i in suburban_driver_count], c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
# Build the scatter plots for rural cities.
plt.scatter(rural_ride_count,
      rural_avg_fare,
      s=[i*10 for i in rural_driver_count], c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
# combining all charts into single chart

plt.subplots(figsize=(10,6))

plt.scatter(urban_ride_count,
      urban_avg_fare,
      s=[i*10 for i in urban_driver_count], c="coral",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count,
      suburban_avg_fare,
      s=[i*10 for i in suburban_driver_count], c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count,
      rural_avg_fare,
      s=[i*10 for i in rural_driver_count], c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")


# incorporating other chart details
plt.title("PyBer Ride-Sharing Data (2019)", fontsize=20)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.xlabel("Total Number of Rides (Per City)", fontsize=12)
plt.grid(True)

plt.legend()
plt.show()
```

```python
# Same chart, but with revised legend

plt.subplots(figsize=(10,6))

plt.scatter(urban_ride_count,
      urban_avg_fare,
      s=[i*10 for i in urban_driver_count], c="coral",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count,
      suburban_avg_fare,
      s=[i*10 for i in suburban_driver_count], c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count,
      rural_avg_fare,
      s=[i*10 for i in rural_driver_count], c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")


# incorporating other chart details
plt.title("PyBer Ride-Sharing Data (2019)", fontsize=20)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.xlabel("Total Number of Rides (Per City)", fontsize=12)
plt.grid(True)

# custom legend
lgnd = plt.legend(fontsize='12', mode='Expanded',
    scatterpoints=1, loc='best', title='City Types')
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)

# Incorporate a text label about circle size.
plt.text(42, 35, "Note: \nCircle size correlates \nwith driver count per city.", fontsize="12")

# Save the figure.
plt.savefig("analysis/Fig1.png", bbox_inches="tight")

plt.show()

```
