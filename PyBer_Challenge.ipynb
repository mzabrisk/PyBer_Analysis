### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "/Users/mattzabriskie/Desktop/Bootcamp/repos/PyBer_Analysis/Resources/city_data.csv"
ride_data_to_load = "/Users/mattzabriskie/Desktop/Bootcamp/repos/PyBer_Analysis/Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_series = pyber_data_df.groupby(['type'], as_index=True).count()['ride_id']
total_rides_series
```

```python
# 2. Get the total drivers for each city type
total_drivers_series = city_data_df.groupby(['type']).sum()['driver_count']
total_drivers_series
```

```python
#  3. Get the total amount of fares for each city type
total_fares_series = pyber_data_df.groupby(['type']).sum()['fare']
total_fares_series
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride_series = total_fares_series/total_rides_series
average_fare_per_ride_series
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver_series = total_fares_series / total_drivers_series
average_fare_per_driver_series
```

```python
#  6. Create a PyBer summary DataFrame.
pyber_summary_df = pd.merge(total_rides_series, total_drivers_series, how='left', on=['type'])
pyber_summary_df = pd.merge(pyber_summary_df, total_fares_series, how='left', on=['type'])
pyber_summary_df['Average Fare per Ride'] = average_fare_per_ride_series
pyber_summary_df['Average Fare per Driver'] = average_fare_per_driver_series
pyber_summary_df.rename(columns={'ride_id': 'Total Rides', 'driver_count': 'Total Drivers', 'fare': 'Total Fares'}, inplace=True)

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df.head()
```

```python
#  8. Format the columns.
pyber_summary_df_formatted = pyber_summary_df.style.format({'Total Rides': '{:,d}',
    'Total Drivers': '{:,d}',
    'Total Fares': '${:,.2f}',
    'Average Far per Ride': '${:,.2f}',
    'Average Fare per Driver': '${:,.2f}'})
pyber_summary_df_formatted

```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
pyber_fares_by_date_df = pyber_data_df.groupby(['type', 'date']).sum()['fare']
#pyber_fares_by_date_df = pyber_data_df[['type','date','fare']].groupby(by=['type', 'date']).sum()['fare']
pyber_fares_by_date_df.head()
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
pyber_fares_by_date_df = pyber_fares_by_date_df.reset_index()
pyber_fares_by_date_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_pivot_df = pyber_fares_by_date_df.pivot(index='date', columns='type', values='fare')
pyber_pivot_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pyber_jan_april_df = pyber_pivot_df.loc['2019-01-01':'2019-04-29']
pyber_jan_april_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
pyber_jan_april_df.index = pd.to_datetime(pyber_jan_april_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pyber_jan_april_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_by_week_df = pyber_jan_april_df.resample('W').sum()
pyber_by_week_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
import numpy as np
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

x_labels = ['Jan\n2019', 'Feb', 'Mar', 'Apr']
#x_axis = np.arange(np.datetime64('2019-01-01'), np.datetime64('2019-05-01'), step = np.timedelta64(30, 'D'))
x_axis = ['2019-01-06', '2019-02-01', '2019-03-01', '2019-04-01']

fig, ax = plt.subplots(figsize=(14,6))
plt.plot(pyber_by_week_df, label=['Rural', 'Suburban', 'Urban'])
plt.legend(loc='best')
plt.title('Total Fare by City Type')
plt.ylabel('Fare ($USD)')
plt.xticks(x_axis, x_labels)
plt.xlim(np.datetime64('2019-01-06'), np.datetime64('2019-04-28'))


plt.savefig('analysis/PyBer_fare_summary.png')

plt.show()





```
